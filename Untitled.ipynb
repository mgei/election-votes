```python
import numpy as np
import http.client
```

```python
conn = http.client.HTTPSConnection("api.srgssr.ch")

payload = ""

authfile = open("key.private", "r")
auth = 'Bearer ' + authfile.read()[0:-1]

headers = { 'accept': "application/json",
            'Authorization': auth}

conn.request("GET", "/polis-api/v2/parties/?lang=de", payload, headers)

res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))
```
